```python
from pyomo.environ import *
import numpy as np
import time
import json
from scipy.interpolate import RegularGridInterpolator
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
latency_file = "latency_head.json"
with open(latency_file) as json_file:
    measurement = json.load(json_file)
EMB = np.arange(4)*256
head = np.array([1,3,6,9,12])
QK = np.array([1,16,32,48,64])
V = np.array([1,16,32,48,64])
MLP = (np.arange(25))*128
data = np.zeros([4,5,5,5,25])
for i in range(3):
    for j in range(5):
        for k1 in range(5):
            for k2 in range(5):
                for l in range(25):
                    e = EMB[i+1]
                    q_h = head[j]
                    q = QK[k1]
                    v = V[k2]
                    if MLP[l]:
                        h = MLP[l]
                    else:
                        h = 1
                    data[i+1,j,k1,k2,l] = measurement['EMB_'+str(e)]['QK_'+str(q_h)+'_'+str(q)]['V_'+str(v)]['MLP_'+str(h)]
latency_look_up_table = RegularGridInterpolator((EMB, head, QK, V, MLP), data) 
```

```python
all_variable_specs = {
    "EMB": [768, 16, 768//16+1],
    "HEAD": [12, 1, 12//1+1],
    "QK": [64, 2, 64//2+1],
    "V": [64, 2, 64//2+1],
    "MLP": [3072, 32, 3072//32+1],
}
```

```python
head_step = 4
step = 8
head_idx_range = np.arange(1, all_variable_specs["HEAD"][2], head_step)
emb_idx_range = np.arange(1, all_variable_specs["EMB"][2], step)
qk_idx_range = np.arange(1, all_variable_specs["QK"][2], step)
v_idx_range = np.arange(1, all_variable_specs["V"][2], step)
mlp_idx_range = np.arange(1, all_variable_specs["MLP"][2], step)
```

```python
mlp_lut[-1, -1] + vandproj_lut[-1, -1, -1] + qk_lut[-1, -1, -1]
```

```python
def regression_test():
    nvit_lats = []
    our_lats = []
    for head_idx in head_idx_range:
        print(head_idx, max(head_idx_range))
        for emb_idx in emb_idx_range:
            for qk_idx in qk_idx_range:
                for v_idx in v_idx_range:
                    for mlp_idx in mlp_idx_range:
                        head = head_idx * all_variable_specs["HEAD"][1]
                        emb = emb_idx * all_variable_specs["EMB"][1]
                        qk = qk_idx * all_variable_specs["QK"][1]
                        v = v_idx * all_variable_specs["V"][1]
                        mlp = mlp_idx * all_variable_specs["MLP"][1]
                        dim_vec = np.array([emb, head, qk, v, mlp])
                        nvit_lat = latency_look_up_table(dim_vec)
                        our_lat = mlp_lut[emb_idx, mlp_idx] + vandproj_lut[head, emb_idx, v_idx] + qk_lut[head, emb_idx, qk_idx]
                        nvit_lats.append(nvit_lat)
                        our_lats.append(our_lat)
    return nvit_lats, our_lats
```

```python
nvit_lats, our_lats = regression_test()
```

```python
nvit_lats, our_lats = np.array(nvit_lats), np.array(our_lats)
```

```python
# Reshape the arrays to be 2D (required for sklearn)
transformed_nvits = transformed_nvits.reshape(-1, 1)
our_lats = our_lats.reshape(-1, 1)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(transformed_nvits, our_lats)

# Predict our_lats values based on transformed_nvits
predicted_lats = model.predict(transformed_nvits)

# Calculate the R^2 value
r2 = r2_score(our_lats, predicted_lats)
print(f"R² value: {r2}")

# Plot the data and the regression line
plt.scatter(transformed_nvits, our_lats, color='blue', label='Data points')
plt.plot(transformed_nvits, predicted_lats, color='red', label='Regression line')
plt.title(f'Linear Regression (R² = {r2:.2f})')
plt.xlabel('Transformed NVITS')
plt.ylabel('Our LATS')
plt.legend()
```

```python
# Filter out elements where transformed_nvits is less than 0.01
mask = transformed_nvits >= 0.01
filtered_nvits = transformed_nvits[mask]
filtered_lats = our_lats[mask]

# Reshape the arrays to be 2D (required for sklearn)
filtered_nvits = filtered_nvits.reshape(-1, 1)
filtered_lats = filtered_lats.reshape(-1, 1)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(filtered_nvits, filtered_lats)

# Predict our_lats values based on filtered_nvits
predicted_lats = model.predict(filtered_nvits)

# Calculate the R^2 value
r2 = r2_score(filtered_lats, predicted_lats)
print(f"R² value: {r2}")

# Plot the data and the regression line
plt.scatter(filtered_nvits, filtered_lats, color='blue', label='Data points')
plt.plot(filtered_nvits, predicted_lats, color='red', label='Regression line')
plt.title(f'Linear Regression (R² = {r2:.2f})')
plt.xlabel('Transformed NVITS')
plt.ylabel('Our LATS')
plt.legend()
plt.show()
```
